# Step 1: Import Required Libraries

In this step, we import the Pandas library, which is used for data loading, cleaning, and analysis.

```python
# Import Pandas Library
import pandas as pd
```

# Step 2: Load the Dataset

The dataset is loaded into a Pandas DataFrame using the read_csv() function.

```python
# Load dataset into DataFrame
df = pd.read_csv("food_coded.csv")
```

# Step 3: Preview the Dataset

Display the first five rows to understand the dataset structure.

```python
df.head()
```

# Step 4: Dataset Information

Check the number of rows, columns, data types, and non-null values.

```python
df.info()
```

```python
df.isnull().sum()
```

# Step 5: Identify Missing Values

Find the number of missing values in each column.

```python
missing_values = df.isnull().sum()
missing_values[missing_values > 0]
```

# Step 6: Handle Missing Values (Numerical Columns)

Replace missing values in numerical columns using the Median because it is less affected by outliers.

```python
numeric_columns = df.select_dtypes(include=['int64', 'float64']).columns

for col in numeric_columns:
    df[col] = df[col].fillna(df[col].median())
```

# Step 7: Handle Missing Values (Categorical Columns)

Replace missing values in categorical columns using the Mode (most frequent value).

```python
object_columns = df.select_dtypes(include=['object']).columns

for col in object_columns:
    df[col] = df[col].fillna(df[col].mode()[0])
```

# Step 8: Verify Missing Values

Confirm that all missing values have been successfully handled.

```python
df.isnull().sum()
```

# Step 9: Check Duplicate Records

Verify whether the dataset contains duplicate rows.

```python
df.duplicated().sum()
```

# Step 10: Verify Data Types

Review the data types of all columns to ensure they are appropriate.

```python
df.dtypes
```

# Step 11: Save the Cleaned Dataset

Export the cleaned dataset as a new CSV file for further analysis.

```python
df.to_csv("food_coded_cleaned.csv", index=False)
```
