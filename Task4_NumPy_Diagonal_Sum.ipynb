## 📌 Objective
The objective of this task is to create a NumPy matrix from user input and calculate the sum of its diagonal elements using a loop.

## 🛠️ Tools Used
- Python
- NumPy
- Jupyter Notebook

## 📚 Concepts Covered
- NumPy Arrays
- User Input
- Loops
- Matrix Operations

```python
import numpy as np
```

## Step 1: Enter Matrix Order

The user enters the size of the square matrix.

```python
n = int(input("Enter the order of the matrix: "))
```

## Step 2: Enter Matrix Elements

The user enters all the elements of the matrix row by row.

```python
matrix = []

print("Enter the matrix elements:")

for i in range(n):
    row = list(map(int, input().split()))
    matrix.append(row)

matrix = np.array(matrix)

print("\nMatrix:")
print(matrix)
```

## Step 3: Calculate Sum of Diagonal Elements

Using a loop, calculate the sum of the main diagonal elements of the matrix.

```python
diagonal_sum = 0

for i in range(n):
    diagonal_sum += matrix[i][i]

print("Sum of diagonal elements =", diagonal_sum)
```
